```python
data['Order Date'] = pd.to_datetime(data['Order Date'])                     ## converting the Strings type data type to the significant value by single values:-
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['OrderDat of Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head(3)
```

```python
data.tail(2)
```

```python
## Q. Find the monthly sales of the company ? :-
```

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month,x = 'Order Month', y =  'Sales', title = "Monthly basis Sales for company")
fig.show()
```

```python
## from the we concluded that jan having very low sales as we have surpassing sales in december.so linearly we are having in profit way.
```

```python
## Q-- Total Analysis of sales by Category( Tech. Office supplies, Furnitures) ?-
```

```python
data.columns
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_category,x = 'Category', y = 'Sales', title = "Monthly basis Category sales  for company")
fig.show()
```

```python
## we can also carry out the the pie chart but in the different fashions:-

sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category,
             names = 'Category',
             values = 'Sales',
             hole = 0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel1)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text = 'Sales analysis by category',title_font = dict(size=20))
fig.show()


```

```python
data.columns
```

```python
## Q -- Total Analysis of monthly Profits :-

profit_by_Months = data.groupby('Order Month')['Profit'].sum().reset_index()
fig = px.bar(profit_by_Months,x = 'Order Month', y = 'Profit', title = "Monthly basis Profits  for company")
fig.show()
```

```python
profit_by_Months    ##From the above we found that in Dec we are having highest profits above all months as compared to september.
```

```python
data.columns
```

```python
## Q--- Total Analysis of  Profits through categories :-
profit_by_Category = data.groupby('Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_Category,x = 'Category', y = 'Profit', title = "Total Analysis of profits through categories")
fig.show()
```

```python
profit_by_Category             ## From the above we showed, that technologies is higher than Office Supplies than the least Furnitures.
```

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
fig = px.pie(profit_by_category,
             names = 'Category',
             values = 'Profit',
             hole = 0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel_r)


fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text = 'Profit analysis by category',title_font = dict(size=20))
fig.show()
```

```python
## Q--- Total Analysis of  Sales & Profits combinely  through Customer Segments  for Company settlements:-
profit_Sales_by_Category = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
profit_Sales_by_Category
```

```python
profit_Sales_by_Segment= data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
color_palette = colors.qualitative.Pastel1

fig = go.Figure()
fig.add_trace(go.Bar(x = profit_Sales_by_Segment['Segment'],
                     y = profit_Sales_by_Segment['Sales'],
                     name = 'Sales',
                     marker_color = color_palette[6]))

fig.add_trace(go.Bar(x = profit_Sales_by_Segment['Segment'],
                     y = profit_Sales_by_Segment['Profit'],
                     name = 'Profit',
                     marker_color = color_palette[0]))
fig.update_layout(title = 'Sales and Profits Of Company Through Segment sections', xaxis_title = 'Customer_Segment', yaxis_title = 'Amount')
```
